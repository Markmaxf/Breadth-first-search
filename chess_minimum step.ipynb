```python
def knightlOnAChessboard(n):
    s=(0,0)
    results=[]
    for i in range(1,n):
        knight_all=[]
        for j in range(1,n):
            knight=[i,j]
            print('-----------------knight{}------------------------'.format(knight))
            if brf_move(s,n,knight) is None:
                knight_all.append(-1)
            else:
                knight_all.append(brf_move(s,n,knight))
        results.append(knight_all)
    return results
        
    
# define brf search function for each knight
def brf_move(s,n,knight):
    level={s:0}
    parent={s:None}
    i=1
    frontier=[s]
    total=n*n-1-1   # total point on board
    while frontier:
        next1=[]         # define next froniter  
        print('----------froniter------------')
        print(frontier)
        for v in frontier:
            print('frontier',v)
            n_point=adj(v,knight,n)         # collect next point
            print('point',v,'adj',n_point)
            for u in n_point:
                if u[0]==n-1 and u[1]==n-1:
                    minlen=max(level.values())+1
                    return minlen
                else:
                    if total==0:
                        return ('66')
                    else:
                        if u not in level.keys():
                            level[u]=i
                            parent[u]=v
                            next1.append(u)
                            total=total-1

        frontier=next1
        i+=1

def adj(v,knight,n):
    next2=[]
    move=[(1,1),(-1,1),(1,-1),(-1,-1)]
    for step in move:

        x=v[0]+step[0]*knight[0]
        y=v[1]+step[1]*knight[1]
        print('step',step,x,y)
        if x>=0 and x<=n-1 and y>=0 and y<=n-1:
            next2.append((x,y))
        x=v[0]+step[0]*knight[1]
        y=v[1]+step[1]*knight[0]
        print('step',step,x,y)
        if x>=0 and x<=n-1 and y>=0 and y<=n-1:
            next2.append((x,y))
    return next2
```
